```python
import unittest

def primes(n):
    a = list(range(n+1))
    a[1] = 0
    lst = []

    i = 2
    while i <= n:
        if a[i] != 0:
            lst.append(a[i])
            for j in range(i, n+1, i):
                a[j] = 0
        i += 1
    return lst

def is_prime(x):
    return (x in primes(x))

def fact(x):
    divisors = primes(x)
    quotient = x
    result = []
    counter = 0
    maxx = max(divisors)
    while divisors[counter] < maxx:
            while (quotient % divisors[counter] == 0):
                result.append(divisors[counter])
                quotient /= divisors[counter]
            counter += 1
    return tuple(result)

def factorize(x):
    """ 
    Factorize positive integer and return its factors.
    :type x: int,>=0
    :rtype: tuple[N],N>0
    """
    
    if type(x) != type(1):
        result = False
        raise TypeError
    elif x < 0:
        raise ValueError
    elif x in [0, 1] or is_prime(x):
        result = (x,)
    else:
        result = fact(x)
    return result


class TestFactorize(unittest.TestCase):
    def test_wrong_types_raise_exception(self):
        ''' проверяет, что передаваемый в функцию аргумент типа float или str \
вызывает исключение TypeError. 
            Тестовый набор входных данных:  'string',  1.5'''
        self.cases = ('string',  1.5)
        for x in self.cases:
            with self.subTest(x=x):
                with self.assertRaises(TypeError):
                    factorize(x)
    
    def test_negative(self):
        ''' проверяет, что передача в функцию factorize отрицательного числа \
вызывает исключение ValueError. 
            Тестовый набор входных данных:   -1,  -10,  -100'''
        self.cases = (-1,  -10,  -100)
        for x in self.cases:
            with self.subTest(x=x):
                with self.assertRaises(ValueError):
                    factorize(x)
    
    def test_zero_and_one_cases(self):
        ''' проверяет, что при передаче в функцию целых чисел 0 и 1, \
возвращаются соответственно кортежи (0,) и (1,). 
            Набор тестовых данных: 0 → (0, ),  1 → (1, )'''
        self.cases = (0, 1)
        for x in self.cases:
            with self.subTest(x=x):
                self.assertEqual((x,), factorize(x))

                   
    
    def test_simple_numbers(self):
        ''' - что для простых чисел возвращается кортеж, содержащий одно данное число. 
            Набор тестовых данных: 3 → (3, ),  13 → (13, ),   29 → (29, )'''
        self.cases = (3, 13, 29)
        for x in self.cases:
            with self.subTest(x=x):
                self.assertEqual((x,), factorize(x))


    
    def test_two_simple_multipliers(self):
        ''' — проверяет случаи, когда передаются числа для которых функция factorize \
возвращает кортеж с числом элементов равным 2. 
            Набор тестовых данных: 6 → (2, 3),   26 → (2, 13),   121 --> (11, 11)'''
        self.cases = {6: (2, 3), 26: (2, 13), 121: (11, 11)}
        for x in self.cases:
            with self.subTest(x=x):
                self.assertEqual(self.cases[x], factorize(x))



    
    def test_many_multipliers(self):
        ''' - проверяет случаи, когда передаются числа для которых функция factorize \
возвращает кортеж с числом элементов больше 2. 
            Набор тестовых данных: 1001 → (7, 11, 13) ,   9699690 → (2, 3, 5, 7, 11, 13, 17, 19)'''
        self.cases = {1001: (7, 11, 13), 9699690: (2, 3, 5, 7, 11, 13, 17, 19)}
        for x in self.cases:
            with self.subTest(x=x):
                self.assertEqual(self.cases[x], factorize(x))

if __name__ == "__main__":
    unittest.main(argv=['ignored_arg'], exit=False)
```
